```python
import matplotlib.pyplot as plt

# Define coffee locations
locations = ["Dublin City Center", "Dublin Suburb", "Cork City", "Galway", "Rural Village"]

# Define coffee prices (Euro)
prices = [3.80, 4.20, 3.50, 2.70, 4.50]

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(locations, prices, color=['brown', 'orange', 'green', 'blue', 'purple'])
plt.xlabel("Location")
plt.ylabel("Price (Euro)")
plt.title("Price of a Cup of Coffee in Ireland (Estimated)")

# Display price on top of each bar
for i, price in enumerate(prices):
  plt.text(i, price + 0.1, f"{price:.2f}", ha='center')

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Display chart
plt.show()
```
